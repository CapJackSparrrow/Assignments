# Q.3.

The arithmetic error occurs when an error is encountered during numeric calculations in Python

ArithmeticError includes:

--> ZeroDivisionError

--> OverflowError

--> FloatingPointError

```python
## example of  ZeroDivisionError

try:
    a = 10 
    a/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
## example of OverflowError

j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

# Q.4.

LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively

LookupError is base class for:

 --> IndexError
 
 --> KeyError

```python
## example of IndexError

try:
    name  = ['James', 'Peter', 'Max', 'Brown']
    print(name[4])
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

```python
## example of KeyError
gender = {'June' : 'female', 'John' : 'male' , 'Kim': 'female'}
    

try:
    gender['Jackey']
   
except KeyError as e:
    print(f'{e}, {e.__class__}')

    
```
