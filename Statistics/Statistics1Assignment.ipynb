# Nominal level

You can categorize your data by labelling them in mutually exclusive groups, but there is no order between the categories

for e.g.

Gender

Marital status

# Ordinal level

You can categorize and rank your data in an order

Difference can't be measured between categories

for e.g.

Language ability (e.g., beginner, intermediate, fluent)

# Interval level

You can categorize and  rank the data

Difference can be measured between data

There is no true zero point

for e.g.

Temperature in Farenheit or Celcius

IQ measurement


# Ratio level

You can categorize, rank, and infer equal intervals between neighboring data points

There is a true zero point

for e.g.

Age ,Height, Weight

# Q.6.

It is important to understand the level of measurement when analyzing data because it affects the types of statistical analyses that can be used, as well as the interpretation and generalization of results

The level of measurement refers to the nature and properties of the variables being measured, which can be classified into four levels: nominal, ordinal, interval, and ratio

For example, 

let's consider a study that examines the relationship between gender and job satisfaction among employees. In this study, the level of measurement for the variable "gender" is nominal, as it is a categorical variable with two or more unordered categories (e.g., male and female). The level of measurement for the variable "job satisfaction" can be either ordinal or interval, depending on how it is measured. If job satisfaction is measured using a Likert scale with five response options ranging from "strongly disagree" to "strongly agree," then it is an ordinal variable, as the categories have a natural order but the distances between them are not equal. If job satisfaction is measured using a scale that has a true zero point, such as a scale that measures the number of days absent from work due to job dissatisfaction, then it is a ratio variable, as it has a true zero point and equal intervals between categories.

# Some common measures of central tendency are:

MEAN: The mean is the average of a set of numbers. It is calculated by adding up all the numbers in the set and dividing by the total number of values. The mean is useful in describing a dataset because it gives an overall idea of where the data cluster around. However, it can be sensitive to outliers.

MEDIAN: The median is the middle value of a set of numbers when they are arranged in order. The median is a useful measure of central tendency when the dataset has extreme values or outliers because it is not affected by extreme values.

MODE: The mode is the most frequent value in a dataset. It is useful when dealing with categorical data, such as the favorite color of a group of people.



# Some common measures of variability are:

RANGE: The range is the difference between the largest and smallest values in a dataset. It gives an idea of the spread of the data, but it is sensitive to outliers.

VARIANCE: The variance is a measure of how much the values in a dataset vary from the mean. It is calculated by taking the average of the squared differences between each value and the mean. Variance can be useful for comparing the variability of two or more datasets.

STANDARD DEVIATION: The standard deviation is the square root of the variance. It is a commonly used measure of variability because it is easier to interpret than the variance. The standard deviation gives an idea of how much the values in a dataset vary from the mean, and is often used to identify outliers
